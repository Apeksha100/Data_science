```python
lst =[1,4,6,7,7,8,97,44]
odd_num = []
for i in lst:
    if i%2!=0:
        odd_num.append(i)
odd_num
        
```

```python
#list comprehension - in single line / inside list 
odd_num = [i for i in lst if i%2==0]
odd_num
```
