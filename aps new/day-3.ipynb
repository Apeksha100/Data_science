```python
import pandas as pd #data manipulation 
import numpy as np #array 
import matplotlib.pyplot as plt # visualization 
from sklearn.model_selection import train_test_split # ml lib - function imported 
from sklearn.linear_model import LinearRegression # class imported
from sklearn.metrics import mean_squared_error, r2_score #to tell the performance of model  
```

```python
df = pd.read_csv('ds2.csv')
df.head() #temp - feat , energy_consumption - target 
```

```python
#feature and target selection 
# feats - represented by x,X - are always multi-dimensional 
# Target - y,Y - not compulsory - in this case only 1 dimensional
X = df[['Temperature (°C)']]
Y = df['Energy Consumption (kWh)']
```

```python
#split ds - 80/20% spliting (train/test) - can also be 70/30%
X_train, X_test, Y_train, Y_test = train_test_split(
X,Y,train_size = 0.8, random_state = 42) # random_state = 42 - gives diff answer if not used , why 42 - fictional story where 42 is answer for everything in universe 
```

```python
# instantiate a class
model = LinearRegression()
```

```python
#train the model 
model.fit(X_train, Y_train)
```

```python
#prediction of this 
Y_pred = model.predict(X_test)
Y_pred
```

```python
# calculate the mean square error
mse = mean_squared_error(Y_test, Y_pred) 
print('The mean Sqaured error is :', mse)
```

```python
# performnace of model - will always be in percentage 
# model evauluation 
r2 = r2_score(Y_test, Y_pred)
print('The r sqaured score is :', r2) - # 61.19% the model is doing well, but the rest can't be generalized 
```

```python
plt.scatter(X_test, Y_test, color = 'blue', label = 'Test Data') #bi-variant analayis  (2 variant )
plt.plot(X_test , Y_pred, color = 'red', label = 'Reg Line ')#line plot
plt.xlabel('Temperature')
plt.ylabel('Energy Consumption')
plt.legend() #the index that shows what is what 
plt.title('Energy Consumption Prediction using SLR') #Simple Linear Regression - 1 feature and 1 target 
plt.show()
```
